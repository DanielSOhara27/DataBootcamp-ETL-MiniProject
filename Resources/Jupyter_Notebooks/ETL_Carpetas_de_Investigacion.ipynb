# Cleaning Carpetas de Investigacion - RAW CSV

This notebook is going to be used to extract the raw dataset into a smaller and more usable one. In other words, we are cleaning the dataset using Pandas, Python and Jupyter Notebooks.

```python
#Import the libraries required to start working
import pandas as pd
```

```python
#Read the csv file
ci = pd.read_csv("../Raw/carpetas-de-investigacion-pgj-cdmx.csv", delimiter=";")
ci.head()
```

```python
# Transform the information into DataFrame
df = pd.DataFrame(ci)
df.head()
```

```python
#Drop the columns that are of no interest
df.drop(columns=["Agencia","Fiscalía","Calle1","Calle2","Mes y año","Latitud","Longitud"], inplace=True)
```

```python
# Show new DataFrame
df.head()
```

```python
#Inspect DataTypes
df.dtypes
```

```python
# Show months in the dataframe
df.Mes.unique()
```

```python
#Show alcaldías
#This will requiere future work
df.Alcaldía.unique() 
```

```python
#Check for null values
df.count()
```

```python
#Drop null values
df.dropna(inplace=True)
```

```python
#Check for null values
df.count()
```

```python
#Save Clean dataframe to a new file
df.to_csv("../Clean/clean_CarpetasInvestigacion.csv", sep= ";", index=False)
```
