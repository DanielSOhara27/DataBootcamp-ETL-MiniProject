# Transformation Carpetas de Investigación - Clean CSV

This notebook is going to be used transform the information contained in the clean_CarpetasInvestigacion file

```python
import pandas as pd
```

```python
#First we import a csv file that contains the alcaldias to be used
alcaldias = pd.read_csv("../Raw/alcaldias.csv", delimiter=";")
alcaldias.head()
```

```python
# We transform the csv file into a pandas dataframe
alc = pd.DataFrame(alcaldias)
alc.head()
```

```python
#We extract the alcaldias and generate a list of unique values
list_alc = list( alc.Alcaldía.unique())
list_alc
```

```python
#We read the file that contains all the information about the investigation files
ci = pd.read_csv("../Clean/clean_CarpetasInvestigacion.csv", delimiter=";")
ci.head()
```

```python
# We transform it into a pandas dataframe
df = pd.DataFrame(ci)
df.head()
```

```python
#We filter only the information of the year 2018
df = df[ df["Año"]==2018 ]
df.head()
```

```python
#Reset the index of the dataframe
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
#From our file we obtain all the unique values
df.Alcaldía.unique()
```

```python
df["Categoría de delito"].unique()
```

```python
#We count the total number of rows in the dataframe
df.count()
```

```python
#We generate a new column that will show True if the Alcaldía is in the alcaldia list, 
#otherwise it shows false

df["Esta_En_Lista"] = df["Alcaldía"].isin(list_alc)

df
```

```python
#Now we homologate the alcaldias name of our dataframe such that it will match those in the
#list
df.loc[ df.Alcaldía=="LA MAGDALENA CONTRERAS" , "Alcaldía" ] = "MAGDALENA CONTRERAS"
df.Alcaldía.unique()
```

```python
# We check the dataframe and we still have alcaldias with different names
df.Esta_En_Lista.unique()
```

```python
#We continue the homologation process
df.loc[ df.Alcaldía=="CUAJIMALPA DE MORELOS" , "Alcaldía" ] = "CUAJIMALPA"
df.Alcaldía.unique()
```

```python
#We check the database and now all the names match
df["Esta_En_Lista"] = df["Alcaldía"].isin(list_alc)

df.Esta_En_Lista.unique()
```

```python
#We drop the temporal column used to homologate the alcaldias names
df.drop(columns=["Esta_En_Lista"], inplace=True)
df.head()
```

```python
#Save dataframe to csv file
df.to_csv("../Clean/after_transform/transformed_CarpetasInvestigacion.csv", sep= ";", index=False)
```
